```python
# Group the data by year and sum the values for each column
summed_data = features.groupby('year').sum()

# Reset the index to make the 'year' column a regular column
summed_data = summed_data.reset_index()

summed_data = summed_data.iloc[:, :13] 

summed_data = summed_data.drop(summed_data.columns[0], axis=1)

summed_data
```

```python
summed_data
```

```python
from sklearn.model_selection import train_test_split

# Splitting features and target variables
X_train, X_test, y_train, y_test = train_test_split(summed_data, yields, test_size=0.2, random_state=42)
```

```python
from torch.utils.data import DataLoader, TensorDataset

# Convert data to tensors
X_train_tensor = torch.Tensor(X_train.values)
y_train_tensor = torch.Tensor(y_train.values)
X_test_tensor = torch.Tensor(X_test.values)
y_test_tensor = torch.Tensor(y_test.values)

# Create TensorDatasets
train_dataset = TensorDataset(X_train_tensor, y_train_tensor)
test_dataset = TensorDataset(X_test_tensor, y_test_tensor)

# Define batch size for data loaders
batch_size = 64

# Create data loaders
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)
```

```python
import torch.nn as nn

class NeuralNetwork(nn.Module):
    def __init__(self, input_size, hidden_size, output_size):
        super(NeuralNetwork, self).__init__()
        self.fc1 = nn.Linear(input_size, hidden_size)
        self.relu = nn.ReLU()
        self.fc2 = nn.Linear(hidden_size, output_size)

    def forward(self, x):
        x = self.fc1(x)
        x = self.relu(x)
        x = self.fc2(x)
        return x  # Output only the first column

# Define the model architecture
input_size = len(summed_data.loc[0])  # Number of input features
hidden_size = 64  # Number of neurons in the hidden layer
output_size = 1  # Number of output neurons (predicting a single value)

model = NeuralNetwork(input_size, hidden_size, output_size)
```

```python
from sklearn.metrics import mean_absolute_error


device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Initialize variables to store predictions and ground truth labels
predictions = []
labels = []

# Iterate over the test DataLoader
for inputs, targets in test_loader:
    # Move inputs and targets to the device (e.g., GPU)
    inputs = inputs.to(device)
    targets = targets.to(device)
    
    # Forward pass
    outputs = model(inputs)
    
    # Convert outputs and targets to numpy arrays
    outputs = outputs.detach().cpu().numpy()
    targets = targets.detach().cpu().numpy()
    
    # Append predicted values and ground truth values to lists
    predictions.extend(outputs[:, 0].tolist())  # Select the first column from the outputs
    labels.extend(targets[:, 0].tolist())  # Select the second column from the targets

# Calculate Mean Absolute Error (MAE)
mae = mean_absolute_error(labels, predictions)

# Print the MAE
print("Mean Absolute Error (MAE):", mae)

print("Labels:", labels)
print("Predictions:", predictions)

```
